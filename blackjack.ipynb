```python
import random 
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

playing = True
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank =  rank
        
        
    def __str__(self):
        return self.rank + " of " + self.suit

class Deck:
    
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
                
    def __str__(self):
        deck_comp = ''
        for card in self.deck:
            deck_comp += '\n' + card.__str__()
        return "The Deck has : "+ deck_comp
        
        
    def shuffle(self):
        random.shuffle(self.deck)
    
    def deal(self):
        single_card = self.deck.pop()
        return single_card
class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
        
    def add_card(self,card):
        #card coming from deal method in deck 
        self.cards.append(card)
        self.value += values[card.rank]
        
        if card.rank == 'Ace':
            self.aces += 1
            
    def adjust_for_ace(self):
            #changing ace value to 1 or 11 according to total cards value        
            while self.value > 21 and self.aces > 0 : 
                self.value -= 10
                self.aces -= 1
            
            
class Chips:
    
        def __init__(self,total=250):
            self.total = total  # This can be set to a default value or supplied by a user input
            self.bet = 0
            
        def win_bet(self):
            self.total += self.bet
        
        def lose_bet(self):
            self.total -= self.bet
            
def take_bet(chips):
    
    while True:
        
        try:
            chips.bet = int(input("How many chips you want to bet"))
        except:
            print("Enter an integer")
        else:
            if chips.bet > chips.total:
                print("Insufficient Chips!!! only {} chips available".format(chips.total))
            else:
                break

def hit(deck,hand):
    
    single_card = deck.deal()
    hand.add_card(single_card)
    hand.adjust_for_ace()

def hit_or_stand(deck,hand):
    global playing #to control while loop
    
    while True:
        x = input("Hit or Stand? Enter h or s")
        if x[0].lower() == 'h':
            hit(deck,hand)
        elif x[0].lower() == 's':
            print("Dealers turn")
            playing = False
        else:
            print("Sorry wrong input enter 'h' or 's' only")
            continue
        break

def show_some(player,dealer):
    
    #show one of dealers card
    
    print("\n Dealer's Hand : " )
    print("First card hidden!")
    print(dealer.cards[1])
    
    #show 2 of players card
    print("\n Players Hand :")    
    for card in player.cards:
        print(card)
    
def show_all(player,dealer):
    #show all dealers card  
    print("\n Dealer's Hand :")    
    for card in dealer.cards:
        print(card)
    #calculate the calue of cards
    print(f'Value of Dealers hand is : {dealer.value}')
    #show all players card  
    print("\n Players Hand :")    
    for card in player.cards:
        print(card)
    print(f'Value of Players hand is : {player.value}')
    
def player_busts(player,dealer,chips):
    print("BUST Player!")
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print("Player won")
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print("Player won")
    chips.win_bet()
        
def dealer_wins(player,dealer,chips):
    print("Dealer won")
    chips.lose_bet()
    
def push():
    print("Dealer and Player tie!! PUSH")
    
#LOGIC 

while True:
    
    print("Welcome to BLACK JACK !!!")
    
    deck = Deck()
    deck.shuffle()
    
    player_hand = Hand() #player hand 
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())
    
    dealer_hand = Hand() #dealer hand
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
    
    player_chips = Chips()#player placing chips to bet
    
    take_bet(player_chips)
    show_some(player_hand,dealer_hand)
    
    while playing:
        #hit or stamd  
        
        hit_or_stand(deck,player_hand)
        
        
        #show cards      
        show_some(player_hand,dealer_hand)
        
        #players busts 
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,player_chips)
            
            break
    
        #if player doesnt bust
    if player_hand.value <= 21:
        
        while dealer_hand.value < 17:
            hit(deck,dealer_hand)
        
        #show all cards
        
        show_all(player_hand,dealer_hand)
        
        #run the game scenes
        if dealer_hand.value > 21:
            dealer_busts(player_hand,dealer_hand,player_chips)
        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_hand,dealer_hand,player_chips)
        elif dealer_hand.value < player_hand.value:
            player_wins(player_hand,dealer_hand,player_chips)
        else:
            push(player_hand,dealer_hand)
        
    print("\nPlayer's winnings stand at",player_chips.total)
    
    # Ask to play again
    new_game = input("Would you like to play another hand? Enter 'y' or 'n' ")
    
    if new_game[0].lower()=='y':
        playing=True
        continue
    else:
        print("Thanks for playing!")
        
        break
        
```
